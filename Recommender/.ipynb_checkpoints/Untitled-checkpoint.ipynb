```python
import tensorflow as tf
import numpy as np
import threading
import time
```

```python
sess = tf.InteractiveSession()
```

```python
gen_random_normal = tf.random_normal(shape=())
```

```python
#사이즈 100짜리 큐 생성.
queue = tf.FIFOQueue(capacity=100,dtypes=[tf.float32],shapes=())
```

```python
#enqueue 정의
enque = queue.enqueue(gen_random_normal)
```

```python
#enque 10번
def add():
    for i in range(10):
        sess.run(enque)
```

```python
# 10개의 쓰레드를 만들고 각각의 쓰레드가 병렬로 add 함수를 비동기적으로 실행한다.
threads = [threading.Thread(target=add,args=()) for i in range(10)]
```

```python
for t in threads:
    t.start()
    
print(sess.run(queue.size()))
time.sleep(0.001)
print(sess.run(queue.size()))
time.sleep(0.001)
print(sess.run(queue.size()))
```
